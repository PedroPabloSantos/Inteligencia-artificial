```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

def busqueda_secuencial_hacia_atras(X, y, umbral_iteraciones_sin_mejora=10):
    # Inicialización
    solucion_actual = list(X.columns)
    n_features = len(solucion_actual)
    resultados = []
    eliminados = []
    añadidos = []
    iteraciones_sin_mejora = 0

    # Función para evaluar condición de parada
    def condicion_de_parada():
        return len(eliminados) == n_features and iteraciones_sin_mejora >= umbral_iteraciones_sin_mejora

    # Ejecución
    while not condicion_de_parada():
        peor_rendimiento = float('inf')
        peor_feature = None

        # Evaluar todas las posibles soluciones temporales
        for feature in solucion_actual:
            if feature not in eliminados:
                solucion_temporal = solucion_actual.copy()
                solucion_temporal.remove(feature)
                rendimiento = evaluar_rendimiento(X, y, solucion_temporal)

                # Guardar el rendimiento de la peor característica
                if rendimiento < peor_rendimiento:
                    peor_rendimiento = rendimiento
                    peor_feature = feature

        # Actualizar la solución actual eliminando la peor característica
        if peor_feature:
            solucion_actual.remove(peor_feature)
            eliminados.append(peor_feature)
            rendimiento_actual = evaluar_rendimiento(X, y, solucion_actual)
            resultados.append((solucion_actual.copy(), len(solucion_actual), rendimiento_actual))

        # Añadir la mejor variable (solo si hay mejora)
        mejor_rendimiento = float('-inf')
        mejor_feature = None

        for feature in X.columns:
            if feature not in solucion_actual and feature not in añadidos:
                solucion_temporal = solucion_actual.copy()
                solucion_temporal.append(feature)
                rendimiento = evaluar_rendimiento(X, y, solucion_temporal)

                if rendimiento > mejor_rendimiento:
                    mejor_rendimiento = rendimiento
                    mejor_feature = feature

        if mejor_feature and mejor_rendimiento > rendimiento_actual:
            solucion_actual.append(mejor_feature)
            añadidos.append(mejor_feature)
            iteraciones_sin_mejora = 0
        else:
            iteraciones_sin_mejora += 1

    # Devolver los resultados
    return resultados

# Aplicación del Algoritmo a los Datos del Titanic
titanic = lectura_titanic()
atributos_discretos_titanic = ['Sex', 'Embarked', 'Alone', 'Deck']
atributos_continuos_titanic = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Initial', 
                               'Age_band', 'Family_Size', 'Fare_cat', 'Title', 'Is_Married']
atributos_titanic = titanic.loc[:, atributos_discretos_titanic + atributos_continuos_titanic]
objetivo_titanic = titanic['Survived']

# Codificación de variables
codificador_atributos_discretos_titanic = OrdinalEncoder()
codificador_atributos_discretos_titanic.fit(atributos_titanic[atributos_discretos_titanic])
atributos_titanic[atributos_discretos_titanic] = codificador_atributos_discretos_titanic.transform(
    atributos_titanic[atributos_discretos_titanic]
)
codificador_objetivo_titanic = LabelEncoder()
objetivo_titanic = codificador_objetivo_titanic.fit_transform(objetivo_titanic)

# Ejecutar la búsqueda secuencial hacia atrás
resultados = busqueda_secuencial_hacia_atras(atributos_titanic, objetivo_titanic)

# Crear una tabla con los resultados
resultados_df = pd.DataFrame(resultados, columns=['MejorSolucionTemporal', 'Tamano', 'Rendimiento'])

resultados_df = resultados_df.sort_values(by='Rendimiento', ascending=False)
# Mostrar la tabla de resultados
print(resultados_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate

# Lectura y procesamiento de datos
def lectura_titanic():
    fichero = pd.read_csv('titanic.csv')    
    return fichero

# Función para evaluar el rendimiento de manera robusta
def evaluar_rendimiento(X, y, features, n_exp=10, cv=5):
    clf = DecisionTreeClassifier(random_state=42)
    resultados = []
    
    for _ in range(n_exp):
        resultados_cv = cross_validate(clf, X[features], y, cv=cv, scoring='balanced_accuracy', return_train_score=False, n_jobs=-1)
        resultados.append(np.mean(resultados_cv['test_score']))
    
    return np.mean(resultados)

# Implementación de la búsqueda secuencial hacia atrás mixta
def busqueda_secuencial_hacia_atras_mixta(X, y, umbral_iteraciones_sin_mejora=10, n_exp=10, cv=5):
    # Inicialización
    solucion_actual = list(X.columns)
    añadidos = []
    eliminados = []
    iteraciones_sin_mejora = 0
    n_features = len(solucion_actual)
    resultados = []
    
    # Función para evaluar condición de parada
    def condicion_de_parada():
        return len(eliminados) == n_features and iteraciones_sin_mejora >= umbral_iteraciones_sin_mejora
    
    # Ejecución
    while not condicion_de_parada():
        # Eliminar la peor variable
        peor_rendimiento = float('inf')
        peor_feature = None
        
        for feature in solucion_actual:
            if feature not in eliminados:
                solucion_temporal = solucion_actual.copy()
                solucion_temporal.remove(feature)
                rendimiento = evaluar_rendimiento(X, y, solucion_temporal, n_exp, cv)
                
                if rendimiento < peor_rendimiento:
                    peor_rendimiento = rendimiento
                    peor_feature = feature
        
        if peor_feature:
            solucion_actual.remove(peor_feature)
            eliminados.append(peor_feature)
            rendimiento_actual = evaluar_rendimiento(X, y, solucion_actual, n_exp, cv)
            resultados.append((solucion_actual.copy(), len(solucion_actual), rendimiento_actual))
        
        # Añadir la mejor variable (solo si hay mejora)
        mejor_rendimiento = float('-inf')
        mejor_feature = None
        
        for feature in X.columns:
            if feature not in solucion_actual and feature not in añadidos:
                solucion_temporal = solucion_actual.copy()
                solucion_temporal.append(feature)
                rendimiento = evaluar_rendimiento(X, y, solucion_temporal, n_exp, cv)
                
                if rendimiento > mejor_rendimiento:
                    mejor_rendimiento = rendimiento
                    mejor_feature = feature
        
        if mejor_feature and mejor_rendimiento > rendimiento_actual:
            solucion_actual.append(mejor_feature)
            añadidos.append(mejor_feature)
            iteraciones_sin_mejora = 0
        else:
            iteraciones_sin_mejora += 1
    
    # Devolver los resultados
    return resultados

# Aplicar la búsqueda secuencial hacia atrás mixta en los datos de entrenamiento
titanic = lectura_titanic()
atributos_discretos_titanic = ['Sex', 'Embarked', 'Alone', 'Deck']
atributos_continuos_titanic = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Initial', 
                               'Age_band', 'Family_Size', 'Fare_cat', 'Title', 'Is_Married']
atributos_titanic = titanic.loc[:, atributos_discretos_titanic + atributos_continuos_titanic]
objetivo_titanic = titanic['Survived']

# Codificación de atributos discretos y objetivo
codificador_atributos_discretos_titanic = OrdinalEncoder()
atributos_titanic[atributos_discretos_titanic] = codificador_atributos_discretos_titanic.fit_transform(
    atributos_titanic[atributos_discretos_titanic]
)

codificador_objetivo_titanic = LabelEncoder()
objetivo_titanic = codificador_objetivo_titanic.fit_transform(objetivo_titanic)

# Ejecutar la búsqueda secuencial hacia atrás mixta
resultados = busqueda_secuencial_hacia_atras_mixta(atributos_titanic, objetivo_titanic)

# Crear una tabla con los resultados
resultados_df = pd.DataFrame(resultados, columns=['MejorSolucionTemporal', 'Tamaño', 'Rendimiento'])

# Mostrar la tabla de resultados
print(resultados_df)
```
